## Chapter 8 Relativistic embedding
The embedding method can be applied to the Dirac equation, and this is important for studying systems containing heavier atoms, where spin-orbit and other relativistic effects come in. Variational embedding for the Dirac equation was developed by [Crampin](http://iopscience.iop.org/article/10.1088/0953-8984/16/49/006?pageTitle=IOPscience) (S. Crampin, J. Phys. Condens. Matter **16** 8875-8889 (2004)), and recent applications include Rashba splitting in surface states on Au(111) by [James and Crampin](http://journals.aps.org/prb/abstract/10.1103/PhysRevB.81.155439) (M. James and S. Crampin, Phys. Rev. B **81** 155439 (2010)), and [Ishida](http://journals.aps.org/prb/abstract/10.1103/PhysRevB.90.235422) (H. Ishida, Phys. Rev. B **90** 235422 (2014)).  
In the following cell we initiate the Relativity class, calling the instance `d`, and setting the speed of light. We can use the actual speed of light in atomic units, $c=137.036$ (numerically the inverse of the fine-structure constant), or if we wish to enhance relativistic effects artificially, use a smaller number.

```python
dirac = dirac.Relativity(c = 137.036)
```

### H-atom in a spherical cavity
The variational embedding method was applied by [Crampin](http://iopscience.iop.org/article/10.1088/0953-8984/16/49/006?pageTitle=IOPscience) to the Dirac equation (S. Crampin, J. Phys. Condens. Matter **16** 8875-8889 (2004)), and one of the demonstrations in this paper was the hydrogen atom in a spherical cavity. The potential is given by
$$V(r)=\begin{cases}-\lambda/r,&r\le R\;(\mbox{region I}),\\\quad V_0,&r> R\; (\mbox{region II}),  \end{cases}$$
where $V_0$ is positive. This is shown in the following figure (from Crampin 2004), <img src=dirac_coulomb.png width=350 height=350/> Region I consists of the spherical cavity with the hydrogenic potential, and region II, where the potential is the constant $V_0$, is replaced by an embedding potential over the spherical surface radius $R$.  
We calculate the bound states for hydrogen, with $\lambda = 1$; the states are labelled by orbital angular momentum $l$ and total angular momentum $j$.

The parameters are set in the following cell, with `R, V, l, j` corresponding to $R,\, V_0,\, l,\, j$. For the values corresponding to table 8.1 (Crampin's table 1) we use `R=3.0, V=10.0, l=0, j=0.5`.

```python
R, V, l, j = 3.0, 10.0, 0, 0.5
dirac.hydrogen(R, V, l, int(round(2*j))) # parameters for H in spherical cavity
eig = []
count = 1
```

In the following cell we calculate the energy eigenvalues for this system, using the basis functions given by eq. (8.46). We are asked to input the number of basis functions $N$ of each kind (large and small components, so that the total number of basis functions is $2N$), and the trial energy $\epsilon$ (the trial mass-energy is given by $w=\epsilon+m_ec^2$). After the lines of input, the program calculates the embedding potential at this trial energy, then the matrix elements, and finally the eigenvalues. These are output in a table giving the energies $E_i$, defined in terms of the mass-energy eigenvalues $W_i$ as $E_i=W_i-m_ec^2$. Note that $N$ should be no greater than 8 or 9, otherwise the results become unstable &ndash; presumably because of linear dependencies.

In the eigenvalue list, the first $N$ values are very large and negative, and correspond to hole energies pushed down by the choice of *balanced* basis functions. With the same parameters, the next two energies agree exactly with Crampin's results for the 1s$_{1/2}$ and 2s$_{1/2}$ states of the confined H-atom.

So that we can study convergence, and the variation of the eigenvalues with trial energy, we can re-run the cell and the results for the lowest two (electron) eigenvalues are accumulated in a table printed at the end of the cell. We also calculate the exact energies, using wave-function matching, further down in this notebook. This enables us to compare the embedding approach with exact values for other states as well as s$_{1/2}$.

```python
N = int(raw_input('Number of basis functions of each type?  '))  # inputs number of basis functions of each type
epsilon = float(raw_input('Trial energy?  '))                    # inputs trial energy
dirac.h_embed_pot(epsilon)                                           # calculates embedding potential at trial energy
dirac.h_matel(N)                                                     # calculates matrix elements
eigenvalues = dirac.h_eigen()                                        # calculates eigenvalues
print 20*'_'
print '{:^6}{:^12}'.format('i','energy E_i')
print 20*'_'
for i in range(2*N):
    print '{:3d}{:+15.6e}'.format(i+1, eigenvalues[i])
add = [N, epsilon, eigenvalues[N], eigenvalues[N+1]]
eig.append(add)
print 37*'_'
print '{:>12}{:1d}{:>8}{:3.1f}'.format('l = ', l,'j = ', j)
print '{:^6}{:^10}{:^20}'.format('basis','trial en.','energies')
print 37*'_'
for i in range(count):
    print '{:3d}{:+10.3f}{:+12.6f}{:+12.6f}'.format(eig[i][0], eig[i][1], eig[i][2], eig[i][3])
count = count + 1
```

#### Exact solution
We can compare these results with the exact results obtained by wave-function matching across the boundary between regions I and II. The function `h_match` in the Relativity class evaluates the ratio of the large to small components of the internal (H-atom) and external (square well) solutions of the Dirac equation at the boundary, $r=R$. This function returns the difference between the ratios $\Delta$ for a given energy $E$, and the energy eigenvalues correspond to $\Delta(E)=0$. We first plot $\Delta$ as a function of $E$ to detemine the required eigenvalue graphically, and then find the eigenvalue more accurately by bisection.  
In the next cell we input the start energy, the finish energy, and the number of energy points for plotting $\Delta(E)$. The calculation of $\Delta(E)$ is then called and the results plotted.

```python
en_start, en_end, number_en = -0.6, -0.3, 100      # start and finish energies, and no. of points for plot
elist = np.linspace(en_start, en_end, number_en)
delta = [dirac.h_match(E) for E in elist]
# Plot delta as a function of E - the zeros correspond to eigenenergies
fig = plt.figure(figsize=(9, 6))
ax = fig.add_subplot(1, 1, 1)
ax.grid(linestyle='-', linewidth=0.2)
ax.axhline(y=1.0, linewidth=0.5, color='blue')
ax.set_xlabel('E (a.u.)', fontsize=14)
ax.set_ylabel('$\Delta$', fontsize=16)
ax.set_title('H-atom in spherical cavity', fontsize=16)
ax.plot(elist, delta, color='red')
```

To refine the energy eigenvalue, we call the function `bisect` from the `scipy.optimize` library. This requires as input, two energies which span the required zero of $\Delta(E)$. The output from `bisect` is the energy at which $\Delta(E)=0$, which we can compare with the embedding results above.

```python
e1, e2 = -0.5, -0.40   # these energies span a zero on the graph
e = bisect(dirac.h_match, e1, e2)
print '%s %1d %s %2.1f %s %11.8f' % ('Exact eigenvalue with l, j =', l,',', j, ' = ', e)
```

### Continuum states in a one-dimensional square well
To illustrate continuum states with the Dirac equation we consider the one-dimensional square well discussed in section 8.1.4, and shown below.
<img src=dirac_well.png width=400 height=400 />
The well constitutes region I, and the regions with $|z|>d/2$ are replaced by a relativistic embedding potential at $z = \pm d/2$. As the spins (either parallel or antiparallel to the $z$-axis) are constant in this problem, the embedding potential defined by eq. (8.14) is a diagonal $2\times 2$ matrix, given by (8.52),
$$\Sigma(W)=\begin{pmatrix}\frac{\textrm{i} k}{W+m_e c^2}&0\\0&\frac{\textrm{i} k}{W+m_e c^2}\end{pmatrix}.$$
Here $k$ is the free-electron wave-vector, given in terms of the mass-energy $W$ by $c\hbar k = \sqrt{W^2-m_e^2c^4}$.
The top-left element relates (say) the spin-parallel large and small components of the two-spinor, and the bottom-right the spin-antiparallel components. We can then treat each spin separately, and use a scalar embedding potential for each spin. The basis set expansion in region I is given by eq. (8.53), with the same expansion coefficients for the large and small components.  
We calculate the local density of states per spin in region I, given by eq. (8.60): this includes the contribution from the small component of the wave-function as well as the large.  
In the next cell we set up the parameters of the square well, its width $d$ and depth $v$; and the parameters for the calculation, the number of basis functions $N$ and the parameter $D$ defining the basis functions (8.54), with $D>d$. We also input the start and finish energies, and the number of energy points at which the local density of states $n_\textrm{I}(E)$ is evaluated. 

```python
d, D, v = 5.0, 6.0, -20.0       # d = width of well, D = parameter defining basis functions, v = depth of well
N = 20                          # number of basis functions
en_start, en_end, number_en = 0.0, 10.0, 1000  # start energy, finish energy, number of energy points
elist = np.linspace(en_start, en_end, number_en)
```

The calculation is called from the next cell, and the Dirac density of states plotted, together with the Schrödinger equation density of states, and the scalar relativistic density of states, in which the electron mass inside the well is given by the relativistic formula eq. (8.80). The comparison of the three is particularly interesting if we reduce the speed of light $c$, set at the top of the notebook, to enhance relativistic effects. (Note that if you want to change $c$, you must initially run the first cell to import the relativity class as `dirac`.)

```python
dirac.square_well(d, D, v)
dirac.sw_matel(N)
rel_dos = [dirac.sw_dos(e) for e in elist]
dirac.sw_nonrel_matel(N)
nonrel_dos = [dirac.sw_nonrel_dos(e, 'n') for e in elist]
semirel_dos = [dirac.sw_nonrel_dos(e, 'y') for e in elist]
fig = plt.figure(figsize=(9, 6))
ax = fig.add_subplot(1, 1, 1)
ax.set_xlabel('E (a.u.)', fontsize=14)
ax.set_ylabel('dos (a.u.)', fontsize=14)
ax.set_title('Density of states in well', fontsize=16)
ax.plot(elist, rel_dos, color='red', label='Dirac dos')
ax.plot(elist, nonrel_dos, color='green', label='Schroedinger dos')
ax.plot(elist, semirel_dos, color='black', linestyle='dashed', label='scalar relativistic dos')
ax.legend(fontsize=12, loc='best')
```
